```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head(5)
```

```python
# Plotting No Of Orders per Month
```

```python
orders_per_month = df.groupby('Month').size()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=orders_per_month.index, y=orders_per_month.values)
plt.title('Number of Orders per Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 5 Customers Based on Money Spent 
```

```python
top_customers = df.groupby('CustomerID')['amount_spent'].sum().nlargest(5)
```

```python
plt.figure(figsize=(10,6))
top_customers.plot(kind='barh', color='green')
plt.title('Top 5 Customers Based on Money Spent')
plt.xlabel('Total Money Spent')
plt.ylabel('Customer ID')
plt.show()
```

```python
#  Distribution of Amount Spent
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['amount_spent'], bins=50, kde=True, color='blue')
plt.title('Distribution of Amount Spent')
plt.xlabel('amount_spent')
plt.ylabel('Frequency')
plt.show()
```

```python
# No of Orders per Day
```

```python
orders_per_day = df.groupby('Day').size()

plt.figure(figsize=(8,6))
sns.barplot(x=orders_per_day.index, y=orders_per_day.values, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Orders Per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# No of orders per Hour
```

```python
orders_per_hour= df.groupby('Hour').size()

plt.figure(figsize=(8,4))
sns.barplot(x=orders_per_hour.index, y=orders_per_hour.values)
plt.title('Orders Per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.show()
```
